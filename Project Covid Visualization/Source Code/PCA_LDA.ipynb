```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
```

```python
df = pd.read_csv('TQHDL-main/4-1-2022.csv')
df.head(5)
```

```python
df.drop(df.index[0], inplace=True, axis=0)
df.drop('Country,Other', inplace=True, axis=1)
df.dropna(inplace=True)
df.head(5)
```

```python
X = df.drop('Continent', axis=1)
y = df['Continent']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
pca = PCA(n_components=2)
pca.fit(X)
X_embed = pca.transform(X)
```

```python
df_pca = pd.DataFrame(data=X_embed, columns=['Component_1st', 'Component_2nd'])
df_pca = pd.concat([df_pca, y], axis=1, join='inner')

df_pca
```

```python
df_pca.to_csv('TQHDL-main/PCA.csv', index=False)
```

```python
lda = LDA(n_components=2)
lda.fit(X, y)
X_embed = lda.transform(X)
```

```python
df_lda = pd.DataFrame(data=X_embed, columns=['Component_1st', 'Component_2nd'])
df_lda = pd.concat([df_lda, y], axis=1, join='inner')
```

```python
df_lda.to_csv('TQHDL-main/LDA.csv', index=False)
```
