**Phân tích hồi qui tuyến tính đơn**

```python
import pandas as pd 
import statsmodels.api as sm
```

```python
column_component = [
 'TotalCases',
 'NewCases',
 'TotalDeaths',
 'NewDeaths',
 'TotalRecovered',
 'NewRecovered',
 'ActiveCases',
 'Serious,Critical',
 'Tot Cases/1M pop',
 'Deaths/1M pop',]
```

```python
data = pd.read_csv("D:\Zoom_HK3_2021\Data visualization\World12345345.csv")
data = data[column_component]
data
```

```python
X = data[["ActiveCases","Serious,Critical","NewCases"]]
Y = data["NewDeaths"]
X = sm.add_constant(X)
```

```python
result = sm.OLS(Y, X).fit()
 
# printing the summary table
print(result.summary())
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = data['TotalCases'].to_numpy().reshape(-1,1)
Y = data['TotalDeaths'].to_numpy().reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
predict =  regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': predict.flatten()})
print(df)
```

**PCA**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_pca = data[["TotalCases","NewCases","TotalDeaths","NewDeaths","TotalRecovered","NewRecovered","ActiveCases","Serious,Critical"]]
scaler.fit(data_pca)
scaler_data = scaler.transform(data_pca)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(scaler_data)

x_pca = pca.transform(scaler_data)
print(x_pca)
```

**Data analysis**

```python
from statsmodels.tsa.stattools import adfuller

def ad_test(dataset):
    dftest = adfuller(dataset,autolag='AIC')
    print("1. ADF : " ,dftest[0])
    print("2. P-value : " ,dftest[1])
    print("3. Num of Lags : " ,dftest[2])
    print("4. Num Of Observcations Used For ADF Regression And Critical Value Calculation : " ,dftest[3])
    print("5. Critical Values : " )
    for key , val in dftest[4].items():
        print("\t",key , " : " , val)
ad_test(data['TotalCases'])
```

**ARIMA**


```python
from statsmodels.tsa.arima_model import ARIMA

train = data.iloc[:-18]
test = data.iloc[-18:]
```

```python
model = ARIMA(train['TotalCases'],order=(1,1,0))
model = model.fit()
model.summary()
```

```python
start = len(train)
end = len(train) + len(test) -1
pred = model.predict(start = start , end = end , typ = 'levels')
pred.index = data.index[start:end+1]
```

```python
pred.plot(legend=True)
test['TotalCases'].plot(legend =True)
```
